### Программы для построения графиков к лабораторным по физике. 3 семестр

#### Первая лабораторная - 3.08.

```python
import matplotlib.pyplot as plt
import pandas as pd
import math
```

```python
# Прямые измерения
U = [0.082, 0.101, 0.108]
T = [1/328, 1/351, 1/367]
L = 0.00001
b = d = 0.002
I = 0.001

sigma = lambda u: I*L/u*b*d
print("Электропроводность: ", sigma(U[0]), sigma(U[1]), sigma(U[2]))
ln = [math.log(sigma(U[2])), math.log(sigma(U[1])), math.log(sigma(U[0]))]
print("ln: ", ln)
print(T)
```

```python
plt.minorticks_on()
plt.grid(True, which='both')
plt.grid(which='minor', color='gray', linestyle=':')
plt.plot(T, ln, color='red', label="ln σ(1/T)")

plt.legend()
plt.ylabel("ln σ")
plt.xlabel("1/T, c")
plt.savefig('lab_5.jpg')
plt.show()
```
