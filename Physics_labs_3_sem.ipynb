```python
l_1 = 0.057
l_0 = 0.025
b =  0.04
R = lambda n : l_1 + (n - 1)*l_0 + b/2
print("Расстояние между осью вращения и центром утяжелителя")
R_i = [R(1), R(2), R(3), R(4), R(5), R(6)]
```

Аппроксимация зависимости T(I) прямой по методу наименьших квадратов.

```python
# прямая имеет вид y = bx + a
def coefficients(X, Y):
    x_middle = np.sum(X) / len(X)
    y_middle = np.sum(Y) / len(Y)
    b = 0 # числитель 
    s = 0 # знаменатель 
    for k in range(len(X)):
        b += (X[k] - x_middle) * (Y[k] - y_middle)
        s += (X[k] - x_middle)**2
    b /= s
    a = y_middle - x_middle*b
    return b, a

t_1 = [16.12, 16.18, 16.30]
t_2 = [16.67, 17.13, 16.96]
t_3 = [18.04, 18.01, 17.98]
t_4 = [19.10, 19.24, 18.92]
t_5 = [20.39, 20.49, 20.38]
t_6 = [21.69, 21.97, 21.73]

T = [(middle(t_1)/n)**2, (middle(t_2)/n)**2, (middle(t_3)/n)**2, (middle(t_4)/n)**2, (middle(t_5)/n)**2, (middle(t_6)/n)**2]
#print("Период колебаний: ", T)

b, a = coefficients(I_i, T)
print(a, b)
i = np.linspace(min(I_i), max(I_i))
t = a + b*i

plt.minorticks_on()
plt.grid(True, which='both')
plt.grid(which='minor', color='gray', linestyle=':')
plt.plot(i, t, color='green', label="T^2(I)")

plt.legend()
plt.ylabel("T^2, c^2")
plt.xlabel("I, H*м")
plt.savefig('lab_1.05.jpg')
plt.show()
```

#### Третья лабораторная - 4.03.

```python
# прямая имеет вид y = bx + a
def coefficients(X, Y):
    x_middle = np.sum(X) / len(X)
    y_middle = np.sum(Y) / len(Y)
    b = 0 # числитель 
    s = 0 # знаменатель 
    for k in range(len(X)):
        b += (X[k] - x_middle) * (Y[k] - y_middle)
        s += (X[k] - x_middle)**2
    b /= s
    a = y_middle - x_middle*b
    return b, a

plt.minorticks_on()
plt.grid(True, which='both')
plt.grid(which='minor', color='gray', linestyle=':')

r_lambda = []
n = [1, 2, 3, 4]
r_lambda.append([1.24**2, 2.06**2, 2.67**2, 3.16**2])
b, a = coefficients(n, r_lambda[0])
i = np.linspace(min(n), max(n))
t = a + b*i
plt.plot(i, t, color='red', label="lambda = 630 нм")

r_lambda.append([1.31**2, 2**2, 2.57**2, 3.03**2])
b, a = coefficients(n, r_lambda[1])
i = np.linspace(min(n), max(n))
t = a + b*i
plt.plot(i, t, color='yellow', label="lambda = 578 нм")

r_lambda.append([1.23**2, 1.97**2, 2.51**2, 2.97**2])
b, a = coefficients(n, r_lambda[2])
i = np.linspace(min(n), max(n))
t = a + b*i
plt.plot(i, t, color='green', label="lambda = 546 нм")
r_lambda.append([1.09**2, 1.81**2, 2.29**2, 2.67**2])
b, a = coefficients(n, r_lambda[3])
i = np.linspace(min(n), max(n))
t = a + b*i
plt.plot(i, t, color='blue', label="lambda = 435 нм")

plt.legend()
plt.ylabel("r^2, мм^2")
plt.xlabel("n")
plt.savefig('lab_4.03.jpg')
plt.show()
```
