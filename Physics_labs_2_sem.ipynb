#### Пятая лабораторная

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.minorticks_on()
plt.grid(True, which='both')
plt.grid(which='minor', color='gray', linestyle=':')

U = [0.01, 1.82, 3.13, 4.02, 4.76, 5.42, 5.93, 6.41, 6.79, 7.15, 7.42, 7.71, 7.95, 8.17, 8.17]  
I = [17.71, 15.02, 13.10, 11.79, 10.68, 9.71, 8.96, 8.25, 7.69, 7.16, 6.76, 6.33, 5.98, 5.65, 5.65]

plt.plot(I, U, color='red', label="U(I)")

plt.legend()
plt.xlabel("I, мА")
plt.ylabel("U, В")
#plt.savefig('lab_5.jpg')
plt.show()
```

Расчет коэффициентов этой зависимости методом наименьших квадратов:

```python
def coef(I, U):
    i_middle = np.sum(I) / len(I)
    u_middle = np.sum(U) / len(U)
    n = 0
    s = 0
    for k in range(len(I)):
        n += (I[k] - i_middle) * (U[k] - u_middle)
        s += (I[k] - i_middle)**2
    n /= s
    m = u_middle - i_middle*n
    return n, m, s, i_middle


r, E, D, middle = coef(I, U)
r = abs(r)
print("Смещение - ЭДС", E)
print("Угловой коэффициент - внутреннее сопротивление", r)

def d(i):
    return (U[i] - (E + r * I[i])) ** 2

d_arr = [d(0), d(1), d(2), d(3), d(4), d(5)]
s_r = np.sqrt((np.sum(d_arr) / D*(len(d_arr) - 2)))
print("СКО для углового коэффициента", s_r)
print("Погрешность сопротивления", 2*s_r)
s_E = np.sqrt((1/len(d_arr) + middle**2/D)*(np.sum(d_arr) / (len(d_arr) - 2)))
print("СКО для смещения", s_E)
print("Погрешность ЭДС", 2*s_E)
```

Расчет мошностей

```python
P_R = [0]*len(I)
P_S = [0]*len(I)
P = [0]*len(I)
n_ = [0]*len(I)
for i in range(len(I)):
    P_R[i] = I[i]*U[i]
    P_S[i] = I[i]**2*r
    P[i] = E*I[i]
    n_[i] =  P_R[i]/P[i]

plt.minorticks_on()
plt.grid(True, which='both')
plt.grid(which='minor', color='gray', linestyle=':')

plt.plot(I, P, color='red', label="P(I)")
plt.plot(I, P_S, color='green', label="P_S(I)")
plt.plot(I, P_R, color='orange', label="P_R(I)")

plt.plot(E/(2*r), max(P_R), 'b.', ms=10, label = "max P_R")
plt.axvline(x = E/(2*r), color='blue', linestyle='--')
print("I* - сила тока при максимальной полезной мощности", E/(2*r))
print("Сопротивление при этом", max(P_R)/(E/(2*r))**2)

plt.legend()
plt.xlabel("I, мА")
plt.ylabel("P, мВт")
plt.savefig('lab_5_P.jpg')
plt.show()
```
