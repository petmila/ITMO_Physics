#### Графики, отражающие законы Рэлея-Джинса, Вина и Планка.

```python
# l - лямбда(длина волны)
# Постоянная Больцмана
k = 1.38054*10**(-23)
# Постоянная Планка
h = 6.64*10**(-34)
c = 299792458
# Постоянная Вина
b = 2.9*10**(-3)
r_рэлей = lambda l, T : (2*math.pi*k*c*T)/l**4
r_планк = lambda l, T : (2*math.pi*h*c*c/l**5)/(np.exp(h*c/(k*T*l)) - 1)
lambda_max = lambda T : b/T
r_global = 0

def R(y, i):
    global r_global
    if i == 0 or i == 800:
        r_global += y
    elif i%2 == 0:
        r_global += 2*y
    else:
        r_global += 4*y

for i in range(801):
    R(r_планк((380 + 0.5*i)*10**(-9), 2700), i)

print(r_global*10**(-9)/6) 
R = (2700**4)*5.67036713*10**(-8)
print(R)
print(r_global*10**(-9)/(6*R))

l = np.linspace(20*10**(-9), 2000*10**(-9), 100)

l_вина_2700 = np.linspace(lambda_max(2700), lambda_max(2700), 100)
r_вина = np.linspace(0, 2*10**13, 100)
l_вина_3300 = np.linspace(lambda_max(3300), lambda_max(3300), 100)

fig = plt.figure(figsize=(10,5))

plt.minorticks_on()
plt.grid(True, which='both')
plt.grid(which='minor', color='gray', linestyle=':')

plt.ylim(0, 2*10**13)
plt.xlim(20*10**(-9), 2000*10**(-9))

plt.plot(l, r_рэлей(l, 2700), color='red', label="Закон Рэлея-Джинса, T=2700K")
plt.plot(l, r_рэлей(l, 3300), color='green', label="Закон Рэлея-Джинса, T=3300K")

plt.plot(l, r_планк(l, 2700), color='purple', label="Закон Планка, T=2700K")
plt.plot(l, r_планк(l, 3300), color='orange', label="Закон Планка, T=3300K")

plt.plot(l_вина_2700, r_вина, color='black', label="Закон Вина, T=2700K")
plt.plot(l_вина_3300, r_вина, color='grey', label="Закон Вина, T=3300K")

plt.xlabel("λ, м")
plt.ylabel("r, Вт/м^3")
plt.legend(prop={"size":11})
```
