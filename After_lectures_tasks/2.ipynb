```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

#### Плотности вероятностей 

```python
r0 = 0.52917720859*10**(-10)
print(r0*(3 + np.sqrt(5)))
r = np.linspace(0, r0*11, 1000)
w_r = lambda r : (4*(r**2)/r0**3)*np.exp(-2*r/r0)
w_r_2 = lambda r : ((2*r0 - r)**2)*r**2/(8*r0**5)*np.exp(-r/r0)

plt.plot(r, w_r(r), label = "1s")
plt.plot(r, w_r_2(r), label = "2s")

plt.axvline(x = r0, color = "red",  linestyle='--', label = "r = ro")
plt.axvline(x = (3 + np.sqrt(5))*r0, color = "green", linestyle='--', label = "r = (3 + √5)r0")

plt.legend()
```
